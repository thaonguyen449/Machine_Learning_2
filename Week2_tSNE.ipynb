```python
import numpy as np 
import pandas as pd
from sklearn.datasets import fetch_openml
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mnist = fetch_openml('mnist_784', version=1, cache=True)
X = mnist.data
y = mnist.target
```

```python
X
```

```python
y
```

```python
X = StandardScaler().fit_transform(X)
X_embedded = TSNE(n_components=2, random_state=0).fit_transform(X)
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(x = X_embedded[:,0], y = X_embedded[:,1], hue = y, palette = sns.hls_palette(10), legend = 'full')
```

```python
pca = PCA(n_components=2)
pca_X = pca.fit_transform(X)
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(x = pca_X[:,0], y = pca_X[:,1], hue = y, palette = sns.hls_palette(10), legend = 'full')
```
